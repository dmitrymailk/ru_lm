```python
from pathlib import Path
import pandas as pd

base_folder = "./models/xglm-4.5B_ru_v10/"
p = Path(base_folder)

prediction_paths = sorted(list(p.rglob("*epoch*.csv")))
once = True
datasets = []
for path in prediction_paths:
    dataset = pd.read_csv(path)
    dataset.drop(
        columns=[
            "gen_config",
        ],
        axis=1,
        inplace=True,
    )
    if not once:
        dataset.drop(
            columns=[
                "prompt",
                "original_answer",
            ],
            axis=1,
            inplace=True,
        )

    dataset.rename(
        columns={
            "finetuned_model_output": dataset["finetuned_model"].iloc[0],
        },
        inplace=True,
    )
    dataset.drop(
        columns=[
            "finetuned_model",
        ],
        axis=1,
        inplace=True,
    )
    datasets.append(dataset)
    once = False
datasets = pd.concat(datasets, axis=1)
datasets
```
