```python
import openai
import pandas as pd

openai.api_key = open("./chat_gpt_token").read()
```

```python
def chat_with_chatgpt(prompt, model="gpt-3.5-turbo"):
    chat_completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
    )

    return chat_completion["choices"][0]["message"]["content"].strip()


user_prompt = "Hello world."
chatbot_response = chat_with_chatgpt(user_prompt)
print(chatbot_response)
```

```python
dataset = pd.read_csv("./models/xglm-4.5B_ru_v10/merged_evaluation.csv")
dataset.head(3)
```

```python
dataset_line = dataset.iloc[0]
import numpy as np

# dataset_line = dataset_line.to_dict()
# dataset_line
model_names = [item for item in dataset_line.index.to_list() if "epoch" in item]
model_names = sorted(model_names)
print(model_names)


def shuffle_names(model_names, debug=False):
    shuffled_names_list = []
    for i in [2, 234, 56]:
        rng = np.random.default_rng(i)
        temp_model_names = model_names[:]
        rng.shuffle(temp_model_names)
        shuffled_names_list.append(temp_model_names)
        if debug:
            print(temp_model_names)
    return shuffled_names_list
```

```python
extract_task = (
    lambda model_name, item: item[model_name]
    .split("\nAssi")[0]
    .split("Human:\n")[1]
    .strip()
)
extract_result = (
    lambda model_name, item: item[model_name].split("Assistant:\n")[1].strip()
)


def evaluation_func(
    item, model_name_1, model_name_2, evaluation_prompt: str, debug=False
):
    task = extract_task(model_name=model_name_1, item=item)

    eval_results = []
    model_names = [
        [model_name_1, model_name_2],
        [model_name_2, model_name_1],
    ]

    eval_prompts = []

    for model_1, model_2 in model_names:
        model_name_1_result = extract_result(model_name=model_1, item=item)
        model_name_2_result = extract_result(model_name=model_2, item=item)

        evaluation_prompt_temp = evaluation_prompt.format(
            task,
            model_name_1_result,
            model_name_2_result,
        )

        eval_prompts.append(eval_prompts)

        if debug:
            print(model_1, model_2)
            print(evaluation_prompt_temp)
            print("-" * 100)

        chatbot_response = chat_with_chatgpt(evaluation_prompt_temp)

        if chatbot_response == "Model_1":
            eval_results.append(model_1)
        elif chatbot_response == "Model_2":
            eval_results.append(model_2)

    if debug:
        print(eval_results)

    result = {"eval_prompts": eval_prompts, "eval_results": eval_results, "winner": ""}

    winner = None

    if (
        eval_results[0] == eval_results[1]
        and eval_results[0] != "Error"
        and eval_results[0] != "Draw"
    ):
        winner = eval_results[0]
    elif eval_results[0] == eval_results[1] and eval_results[0] == "Draw":
        winner = model_name_1
    else:
        winner = eval_results[0]

    result["winner"] = winner

    return result
```
